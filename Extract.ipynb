```python
SYMBOL ="{}([]).,:;+-*&\|<>=~%$1234567890_!?#@abcdefghijklmnopqrstuvwxyz /'\n "

def is_none(x):
    if x ==" ":
        return 0
    else:
        return 1
    
def count_emoji(y):
    if (is_none !=0):
        txt = re.sub(r'[\w]', '', y)
        txt = txt.replace("'"," ")
        rem = str.maketrans('','',SYMBOL)
        txt = txt.translate(rem)
        if len(txt) != 0: 
            return len(txt)
        else: 
            return 0
    else:
        return 0
    
def is_prv(z):
    if z == True:
        return 1
    else:
        return 0

def has_pp(b):
    if b[:18] == "https://instagram.":
        return 0
    else:
        return 1
    
    
def emoji_str(y):
    if (is_none !=0):
        txt = re.sub(r'[\w]', ' ', y)
        rem = str.maketrans('','',SYMBOL)
        txt = txt.translate(rem)
        if len(txt) != 0: 
            return txt
        else: 
            return " "
    else:
        return " "
    
def bio_ext(L1):
    L2 = emoji_str(L1)
    if L2 != " ":
        for i in L1:
            for j in L2:
                if i == j:
                    L1 = L1.replace(i,"")
                    rem = str.maketrans('','','\n@!#$%^&*(){}[];:><.,/?-|_0123456789')
                    L1 = L1.translate(rem)
        
        L1 = L1.split()
        return L1
    else:
        return [ ]
```
